## Random dot stereogram demo  

This is a demo for the Week 6 sections on creating random dot stereogram. This is sample Python code for generating a stereogram with a pop-out square in the center.

### Import some python packages

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### Create random cyan and red dots

```python
# Create cyan dots
cyan_x_coordinates = np.random.rand(6000) # uniform distribution from 0 
cyan_y_coordinates = np.random.rand(6000)

# Think carefully about what shift you want - do you want both a horizontal and vertical shift, or just one?
# If it's just one type of shift, which direction should you shift it?
# Hint: Think about how your two eyes are shifted.
# You may need to comment out one of the following lines depending on your answer.
# Create red dots
red_x_coordinates = cyan_x_coordinates + 0.006  # shift to the right
red_y_coordinates = cyan_y_coordinates - 0.004 # shift down
```

### Visualize random dots

```python
fig = plt.gcf()
fig.set_size_inches((15,10))
plt.scatter(cyan_x_coordinates, cyan_y_coordinates, color='c')
plt.scatter(red_x_coordinates, red_y_coordinates, color='r')
plt.xlim([0,1])
plt.ylim([0,1])
plt.axis('off')
```

### Now we're going to move red dots in a center square to create a "pop-out" effect

```python
# Make a 0.3 by 0.3 square in the center
center_indices = (red_x_coordinates >= 0.35) & (red_x_coordinates <= 0.65) \
    & (red_y_coordinates >= 0.35) & (red_y_coordinates <= 0.65)
red_x_coordinates[center_indices] += 0.015 # shift center right
```

```python
fig = plt.gcf()
fig.set_size_inches((15,10))
plt.scatter(cyan_x_coordinates, cyan_y_coordinates, color='c')
plt.scatter(red_x_coordinates, red_y_coordinates, color='r')
plt.xlim([0,1])
plt.ylim([0,1])
plt.axis('off')
```
